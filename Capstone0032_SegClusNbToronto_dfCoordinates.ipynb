### This is the part 2 of the Week-3 Assignment: Dataframe of coordinates.

Build a dataframe of the postal code of each neighborhood along with the borough name and neighborhood name, and the latitude and the longitude coordinates of each neighborhood.

In part 1 of the Week-3 Assignment, we built a dataframe of the postal code of each neighborhood along with the borough name and neighborhood name. We repeat steps here, with output simplified. 

```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_url = pd.read_html(url) # install lxml
df_url # list of a lot of info. [180 rows x 3 columns] and more
df_brnb = df_url[0] # this has what we want: index (0-179), 3 columns (Postal Code, Borough, Neighbourhood)
df_brnb.columns = ["PostalCode", "Borough", "Neighborhood"] #  ["Postal Code", "Borough", "Neighbourhood"] 
df_brnb = df_brnb[df_brnb.Borough != 'Not assigned'] # 103 rows × 3 columns
df_brnb.Neighborhood[df_brnb.Neighborhood=='Not assigned'] = df_brnb.Borough[df_brnb.Neighborhood=='Not assigned']
df_brnb = df_brnb.groupby(['PostalCode','Borough'],as_index=False)['Neighborhood'].apply(', '.join)
print(df_brnb.shape)
df_brnb
```

Read into a dataframe a csv file that has the geographical coordinates of each
postal code: http://cocl.us/Geospatial_data. For convenience, we rename column 'Postal Code' read in as 'PostalCode'. 

```python
df_coor = pd.read_csv("http://cocl.us/Geospatial_data")
df_coor.columns = ['PostalCode', 'Latitude', 'Longitude']
print(type(df_coor))
print(df_coor.shape)
print(df_coor.index)
print(df_coor.columns)
df_coor.head()
```

```python
print("df_brnb.shape, df_brnb.columns:")
print(df_brnb.shape)
print(df_brnb.columns)
print("df_coor.shape, df_coor.columns:")
print(df_coor.shape)
print(df_coor.columns)
```

*`Direction `* Use the csv file to create the following dataframe:

    The dataframe has number index and 5 columns (PostalCode, Borough, Neighborhood, Latitude, Longitude). 
    
Dataframe df_brnb has 3 columns: PostalCode, Borough, Neighborhood. Dataframe df_coor has 3 columns: PostalCode, Latitude, Longitude. Both have 103 rows. Columns df_brnb.PostalCode and column df_coor.PostalCode are of the same data type 'object' after we check df_brnb.dtypes and df_coor.dtypes.

```python
df_brnb.dtypes, df_coor.dtypes, df_brnb.PostalCode.dtypes == df_coor.PostalCode.dtypes
```

Merging/Joining df_brnb and df_coor gives what we want. 

```python
df_bnco = pd.merge(df_brnb, df_coor, on='PostalCode')
print(df_bnco.shape)
df_bnco
```

Now we need to make sure that df_bnco is indeed the dataframe we want. 
We will check no PostalCode is missing in either df_brnb from web or df_coor from csv file. 
This can be verified by comparing merge(df_brnb, df_coor) and merge(df_coor, df_brnb). 
Any missing PostalCode in df_brnb or df_coor will result in different numbers of rows in df_bnco and in, df_cobn.


```python
df_cobn = pd.merge(df_coor, df_brnb, on='PostalCode')
df_cobn
print(all(df_bnco.PostalCode   == df_cobn.PostalCode))
print(all(df_bnco.Borough      == df_cobn.Borough))
print(all(df_bnco.Neighborhood == df_cobn.Neighborhood))
print(all(df_bnco.Latitude     == df_cobn.Latitude))
print(all(df_bnco.Longitude    == df_cobn.Longitude))
```

The above True values show that both df_bnco and df_cobn are essentially the same. Now it is sure that every PostalCode in df_brnb and every PostalCode in df_coor are one-to-one mapped: No PostalCode is missing in either dataframe.
