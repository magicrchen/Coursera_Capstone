### This is the part 1 of the Week-3 Assignment: Build the code to scrape the Wikipedia page


```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_url = pd.read_html(url) # install lxml
df_url # list of a lot of info. [180 rows x 3 columns] and more
df = df_url[0] # this has what we want: index (0-179), 3 columns (Postal Code, Borough, Neighbourhood)
df
```

```python
print(type(df))
print(df.shape)
print(df.index)
print(df.columns)
```

*`Direction 1: `* The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood. 

All columns have no space now. So a column can be referred to in two ways, such as df\["PostalCode"\] or simply df.PostalCode. And column-names are in American spelling.

```python
print(df.columns)
df.columns = ["PostalCode", "Borough", "Neighborhood"] #  ["Postal Code", "Borough", "Neighbourhood"] 
print(df.columns)
print(df.describe())
```

*`Direction 2:`* Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
df = df[df.Borough != 'Not assigned'] # 103 rows Ã— 3 columns
print(df.shape)
df
```

*`Direction 3: `* (This was step 4 direction in assignhment.) If a clel has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. 

```python
df.Neighborhood[df.Neighborhood=='Not assigned'] = df.Borough[df.Neighborhood=='Not assigned']
df
```

*`Direction 4: `* (This was step 3 direction in assighnment.) More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia 
page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park.
These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table. Actually in the present version of Wikipedia file, M5A is listed only once by checking `df[df.PostalCode == 'M5A']`. And we reset index.

```python
df = df.groupby(['PostalCode','Borough'],as_index=False)['Neighborhood'].apply(', '.join)
df
```
